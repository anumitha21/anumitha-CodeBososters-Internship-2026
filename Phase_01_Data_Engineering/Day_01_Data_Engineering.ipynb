```python
#score analysis
math_highest_score = df["math_score"].max()
math_lowest_score = df["math_score"].min()
math_average_score = df["math_score"].mean()
programming_average_score = df["programming_score"].mean()
attendance_average_score = df["attendance_percentage"].mean()

```

```python
#printing score analysis
print("Math Highest Score:", math_highest_score)
print("Math Lowest Score:", math_lowest_score)
print("Math Average Score:", math_average_score.round(2))
print("Programming Average Score:", programming_average_score)
print("Attendance Average Score:", attendance_average_score)
```

```python
#gender-wise analysis
gender_avg = df.groupby("gender")["total_score"].mean().round(2)
gender_avg
```

```python
#top performers
top_students = df.sort_values(by="total_score",ascending=False).head(5)
toppers=top_students[["name", "department", "total_score"]]
toppers
```

```python
# attendance <75 - condt
low_attendance = df[df["attendance_percentage"] < 75]

if low_attendance.empty:
    print("All students have attendance above 75%")
else:
    for index, row in low_attendance.iterrows():
        print("Name :", row["name"])
        print("Department :", row["department"])
        print("Attendance :", row["attendance_percentage"])
        print(" ")
```
